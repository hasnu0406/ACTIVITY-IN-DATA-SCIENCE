1. Write python code for indexing rows and columns using iloc or loc method based on the following conditions:

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("bank_train.csv")
df
```

1. select the rows where clients have subscribed to a deposit?

```python
df.loc[df['deposit']=='yes']
```

2. select the rows where clients who have not subscribed to a deposit?

```python
df.loc[df['deposit']=='no']
```

3. select the rows where clients who have subscribed to a deposit either have a housing or a personal loan?

```python
df.loc[(df['deposit']=='yes')&((df["housing"]=='yes')|(df["loan"]=='yes'))]
```

4. select the rows where clients with secondary education who have not subscribed to a deposit?

```python
df.loc[(df['education']=='secondary')&(df['deposit']=='no')]
```

1. select the rows where clients with primary education have subscribed to a deposit?

```python
df.loc[(df['education']=='primary')&(df['deposit']=='yes')]
```

6. select the rows where clients who have subscribed to a term deposit as an outcome of the successful marketing campaign? 

```python
df.loc[(df['deposit']=='yes')&(df['poutcome']=='success')]
```

7. select the rows where unemployed clients who have not subscribed to deposit?

```python
df.loc[(df['job']=='unemployed')&(df['deposit']=='no')]
```

8. Select columns 'education' and 'balance' where age is less than or equal to 30:

```python
age_fil=df[df['age']<=30][['education','balance']]
age_fil
```
